```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.formula.api as stm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
sleep_health_and_lifestyle_df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

sleep_health_and_lifestyle_df.head()
```

```python
standardized_based_on_occupation = sleep_health_and_lifestyle_df
standard = standardized_based_on_occupation.groupby(['Occupation', 'Age'] )
for i, row in standardized_based_on_occupation.iterrows(): 
    s = (row['Sleep Duration'] - standard.get_group(row['Occupation', 'Age'])['Sleep Duration'].mean()).astype(float)
    standardized_based_on_occupation.at[i, 'standardize sleep'] = s/(standard.get_group(row['Occupation', 'Age'])['Sleep Duration'].std())
```
