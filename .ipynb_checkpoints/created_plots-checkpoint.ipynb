```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.formula.api as stm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
sleep_health_and_lifestyle_df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

sleep_health_and_lifestyle_df.head()
```

```python
group_occupation = sleep_health_and_lifestyle_df.groupby('Occupation')
group_age = sleep_health_and_lifestyle_df.groupby('Age')
group_gender = sleep_health_and_lifestyle_df.groupby('Gender')
group_steps = sleep_health_and_lifestyle_df.groupby('Daily Steps')
group_phy_act = sleep_health_and_lifestyle_df.groupby('Physical Activity Level')
graph, plots = plt.subplots(4, figsize=(20,20), sharex=True)

plots[0]. barh(list(group_occupation.groups.keys()), group_occupation['Stress Level'].mean(), label='Occupation')
plots[1]. barh(list(group_age.groups.keys()), group_age['Stress Level'].mean(), label='Age group')
plots[2]. barh(list(group_gender.groups.keys()), group_gender['Stress Level'].mean(), label= 'Gender')
plots[3]. barh(list(group_phy_act.groups.keys()), group_phy_act['Stress Level'].mean(), label = 'Physical activity')

for i in range(4):
    plots[i].legend()
    
```

The above bar plots show mean stress level for different compnents such as occupation, age, gender, and physical activity. It seems that certain age, gender, and occupation do contribute to stress. For example, Doctors and sales representative have the highest amount of stress level. Also, Male stress levels are higher than female and older people tend to lower levels of stress. Certain amount of physical activity reduces stress levels significantly.

Therefore, it is important to understand where the stress in a person's life comes from. Understanding which component contributes to most amount of stress can help us further understand relationship between sleep patterns and factors such as age, gender, occupation, etc.
